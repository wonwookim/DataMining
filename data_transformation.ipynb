##Data Transformation

* 데이터는 먼저 분석가가 쉽게 분석할 수 있도록 조직되고 구조화되어야 함

* 데이터 변환은 raw data를 수정, 계산, 분리 및 분석 가능한 데이터 모델로 결합하는 프로세스

###Data Transformation:Revising(수정)
* 데이터를 수정하면 값이 정확하고 의도된 사용을 지원하는 방식으로 구성



* Type
  * Data Cleansing
  * Format Revision/Conversion (교정)
  * Key Restructuring (핵심 구조 조정)
  * Deduplication (중복 제거)
  * Data Validation (데이터 타당성)
  * Removing Unused and Repeated Columns(불필요 요소 제거)

###Data Transformation: Computing(계산을 통한 변환)
  * 기존 데이터에서 새로운 데이터 값을 계산하는 것은 비율, 요약, 통계 및 기타 중요한 수치를 계산하는 것

  * Type
    * Derivation (파생)
    * Summarization(요약)
    * Pivoting (피벗)
    * Sorting, ordering, and indexing(정렬, 정렬 및 인덱싱)
    * Scaling, standardization, and normalization(확장, 표준화 및 표준화)
    * Vectorization(벡터화)

###Data Transformation: Separating(분리)
  * Separating은 구성 요소를 나누는 것
    * 데이터 값은 결합되어있을 수 있지만 세분화된 분석을 위해서는 분리해야 할 수 있음
  * Type
    * Splitting
    * Filtering


###Data Transformation: Combining
* Combining은 서로 다른 테이블의 값을 합치는 것을 의미
* Type
  * Joining
  * Merge

####Merge

```python
import pandas as pd
df1 = pd.DataFrame({'data1' : [0,1,2,3,4,5,6], 'key': ['b','b','a','c','a','a','b']})
df2 = pd.DataFrame({'data2' : [0,1,2], 'key' : ['a','b','d']})

pd.merge(df1,df2, on = 'key') # 왼쪽은 기준이 될 데이터, 2번째는 붙일 데이터, 
# on은 어떤 열 기준으로 붙일지 결정 -> key 기준으로 둘 다 가지고 있는 값만 보여줌줌
```

```python
pd.merge(df1, df2, how = 'outer') # how = 'outer' 사용하면 중복되지 않는 값들도 보여줌
```

```python
pd.merge(df1, df2, how = 'left')
```

```python
pd.merge(df1, df2, how = 'right')
```

```python
pd.merge(df1, df2, how = 'inner') #default 값
```

```python
left = pd.DataFrame({'key' : ['apple', 'apple', 'bar'], 
                     'left_' : [1,2,3],
                     'num' : ['one','two','one']})
right = pd.DataFrame({'key' : ['apple', 'apple', 'bar','bar'],
                      'right_' : [4,5,6,7],
                      'num' : ['one','one','one','two']})
```

```python
left1 = pd.DataFrame({'key' : ['a','b','a','a','b','c'], 'value' : [0,1,2,3,4,5]})
right1 = pd.DataFrame({'group_val' : [3.5,7.0]}, index = ['a','b'])
```

```python
pd.merge(left1, right1, left_on = 'key', right_index= True) #다른 것을 기준으로 
# merge를 할 때 left_on은 앞에 데이터프레임에서 어떤것을 비교할지? 
# right_index는 뒤 데이터 프레임에서 index 값으로 비교할 것인지 선택택
```

####Concat

```python
df1 = pd.DataFrame({'A':['A0','A1','A2'],
                    'B' : ['B0','B1','B2'],
                    'C' : ['C0','C1','C2'],
                    'D' : ['D0','D1','D2']},
                   index = [0,1,2])

df2 = pd.DataFrame({'A':['A3','A4','A5'],
                    'B' : ['B3','B4','B5'],
                    'C' : ['C3','C4','C5'],
                    'D' : ['D3','D4','D5']},
                   index = [3,4,5])

df3 = pd.DataFrame({'A':['A6','A7','A8'],
                    'B' : ['B6','B7','B8'],
                    'C' : ['C6','C7','C8'],
                    'D' : ['D6','D7','D8']},
                   index = [6,7,8])
```

```python
frames = [df1,df2,df3]
frames
```

```python
pd.concat(frames)
```

```python
result = pd.concat(frames, keys = ['x', 'y','z']) 
#keys는 합쳐진 데이터 프레임 개수만큼 사용해야함 그 이상 그 이하 x
result
```

```python
result.loc['x']
```

```python
df4 = pd.DataFrame({'B' : ['B2', 'B3','B6'],
                    'D' : ['D2', 'D3', 'D6'],
                    'F' : ['F2', 'F3',' F6']},
                   index = [2, 3, 6])
df4
```

```python
result = pd.concat([df1,df4], ignore_index = True) #join이라는 변수로 합집합 교집합 변경 가능능
```

```python
result
```

####Reshaping


```python
data = pd.DataFrame({'state' : ['Ohio', 'Colorado'],
                     'one' : [0,3],
                     'two' : [1,4],
                     'three' : [2,5]})

data = data.set_index('state')
data
```

```python
data = data.reset_index()
data
```

```python
data.columns.names = ['number']
data = data.set_index('state')
data
```

```python
result = data.stack()
result
```

```python
result.unstack()
```

```python
result.unstack('state')
```
